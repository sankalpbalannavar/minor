```python
import os
import numpy as np
from nibabel.testing import data_path
import matplotlib.pyplot as plt
import nibabel as nib
```

```python
# Get a list of all .nii files in the data directory
data_dir = os.path.join(data_path, "C:\\Users\\HPC\\Desktop\\minor files\\ImageCHD_dataset\\images with label\\images")
file_list = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.nii.gz')]
print(file_list)
```

```python
# Load each file into a NumPy array and append it to a list
data_list = []
for file in file_list:
    print("Added.")
    img = nib.load(file)
    data = img.get_fdata()
    data_list.append(data)
```

```python
i=0
for file in file_list:
    plt.imshow(data_list[i][:,:,100])
    plt.savefig('img'+str(i)+'.png')
    plt.show()
    i=i+1
```
